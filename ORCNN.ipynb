```python
# Объявляем абсолютные пути к нашим данным.
train_dir = '/Users/maksim/PycharmProjects/OrnamentRecognitionCNN/data/train'
valid_dir = '/Users/maksim/PycharmProjects/OrnamentRecognitionCNN/data/valid'
test_dir = '/Users/maksim/PycharmProjects/OrnamentRecognitionCNN/data/test'

# Указываем размер одной группы - указываем 16, т.к. дата-сет имеет не большой объём
batch_size = 16
num_epochs = 30
```

 - Создание трансформаторов данных:
 В связи с тем что наши картинки имеют разное разрешение, нам нужно стандартизировать их, а так же выполнить базовые преобразования нормализации и получить набор тензоров из наших картинок, для этого идеально подойдёт модуль transforms из torchvision.

```python
# Сразу собираем все три конфигурации в один словарь для удобства работы
data_transforms = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)), # 1. Меняем разрешение
        transforms.RandomHorizontalFlip(), # 2. Случайно отражаем наше изображение с вероятностью 0,5
        transforms.RandomRotation(degrees=10), # 3. Случайно поворачиваем изображение на угол от -10 до 10 градусов
        transforms.ToTensor(), # 4. Приводим картинки к тензорам
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]) # 5. Нормализация тензоров
    ]),
    'valid': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'test': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
}
```

Пункты 1 и 4, я описал выше, это необходимо для обучения и работы нашей сети. Пункты 2 и 3, разнообразят наши данные слегка изменяя их. 5 пункт или нормализацию мы определяем исходя из значений ImageNet (значения 0.485, 0.456 и 0.406 являются средними значениями пикселей изображений в дата-сете ImageNet, а значения 0.229, 0.224 и 0.225 - их стандартными отклонениями), это позволит привести все значения к диапазону от -1 до 1, что помогает ускорить сходимость обучения модели.

 - Загружаем данные:
    Стоит отметить что при загрузке данных наши классы определятся автоматически, благодаря нашему разбиению классов на папки

```python
# Загружаем данные из папок и применяем к ним нашу конфигурацию, так же определяем их в словаре
image_datasets = {
    'train': torchvision.datasets.ImageFolder(train_dir, transform=data_transforms['train']),
    'valid': torchvision.datasets.ImageFolder(valid_dir, transform=data_transforms['valid']),
    'test': torchvision.datasets.ImageFolder(test_dir, transform=data_transforms['test'])
}

# Создаём dataloaders с указанным ранее batch_size и shuffle=True - для того чтобы перемешать наши картинки
dataloaders = {
    'train': DataLoader(image_datasets['train'], batch_size=batch_size, shuffle=True),
    'valid': DataLoader(image_datasets['valid'], batch_size=batch_size, shuffle=False),
    'test': DataLoader(image_datasets['test'], batch_size=len(image_datasets['test']), shuffle=False)
}
```

#### Определение метрик

При определении метрик мы можем взять стандартный Accuracy, однако он может быть неинформативен для нас так как мы имеем не сбалансированные классы.

Precision(точность) и Recall(полнота), а так же F1-score, дабы отследить ложно-положительные и ложно-отрицательные результаты классификации, и среднее между ними. А так же Confusion Matrix для понимания какие классы "путаются" больше всего.

```python
# Так же объявляем наши метрики в словаре. Задаём количество наших классов и режим, в данном случае одна картинка не может принадлежать нескольким классам и классов больше двух (multiclass).
metrics = {
    'accuracy': torchmetrics.Accuracy(task='multiclass', num_classes=len(image_datasets['train'].classes)),
    'precision': torchmetrics.Precision(num_classes=len(image_datasets['train'].classes), average='weighted', task='multiclass'),
    'recall': torchmetrics.Recall(num_classes=len(image_datasets['train'].classes), average='weighted', task='multiclass'),
    'f1_score': torchmetrics.F1Score(num_classes=len(image_datasets['train'].classes), average='weighted', task='multiclass'),
    'confusion_matrix': torchmetrics.ConfusionMatrix(task='multiclass', num_classes=len(image_datasets['train'].classes)),
}

for metric in metrics.values():
    metric.to(device)
```

#### Построение модели

Затем приступим к реализации нашей ResNet50:

```python
resnet = models.resnet50(pretrained=True).to(device)
```

Замораживаем веса всех слоев, чтобы не потерять информацию, которую уже получили при предобучении.

```python
for param in resnet.parameters():
    param.requires_grad = False
```

Изменяем последний слой для нашего количества классов (3 в данном случае, но мы определяем кол-во классов по набору данных, что обеспечит гибкость при дальнейшем расширении модели)

```python
num_features = resnet.fc.in_features
resnet.fc = torch.nn.Linear(num_features, len(image_datasets['train'].classes))
```

Определяем оптимизатор и функцию потерь. Так как наша модель предобучена, уменьшим стандартную скорость обучения в 2 раза, до 0,0005

```python
criterion = nn.CrossEntropyLoss().to(device)
optimizer = optim.Adam(resnet.parameters(), lr=0.0005)
```

```python
# Модификация модели

num_samples = len(dataloaders['train'].dataset)
class_freqs = [0] * len(dataloaders['train'].dataset.classes)
for data in dataloaders['train'].dataset:
    class_freqs[data[1]] += 1
class_weights = [num_samples / (len(dataloaders['train'].dataset.classes) * freq) for freq in class_freqs]

weights = [class_weights[data[1]] for data in dataloaders['train'].dataset]
sampler = WeightedRandomSampler(weights, num_samples)

class_weights
```

```python
def modify_resnet():
    global criterion, optimizer, class_weights, device, resnet, dataloaders, sampler, batch_size, image_datasets
    criterion = nn.CrossEntropyLoss(weight=torch.FloatTensor(class_weights)).to(device)
    optimizer = optim.Adam(resnet.parameters(), lr=0.0002)

    dataloaders.update({
        'train': DataLoader(image_datasets['train'], batch_size=batch_size, sampler=sampler)
    })

modify_resnet()
```

Приступаем к обучению модели:

 - Для начала создадим массивы наших метрик, по ним мы в дальнейшем будем строить графики

```python
train_metrics_history = {
    'accuracy': [],
    'precision': [],
    'recall': [],
    'f1_score': [],
    'confusion_matrix': []
}

valid_metrics_history = {
    'accuracy': [],
    'precision': [],
    'recall': [],
    'f1_score': [],
    'confusion_matrix': []
}

train_loss_history = []
valid_loss_history = []

# Функция сброса метрик
def clear_metric():
    global train_metrics_history, valid_metrics_history, train_loss_history, valid_loss_history
    for i in train_metrics_history.values():
        i.clear()

    for i in valid_metrics_history.values():
        i.clear()

    train_loss_history = []
    valid_loss_history = []
```

 - Далее приступаем к обучению нашей модели, не забывая считать и выводить промежуточные результаты метрик

```python
def learn(model_, optimizer_, criterion_):
    global dataloaders, metric, valid_metrics_history, train_metrics_history, train_loss_history, valid_loss_history
    for epoch in range(num_epochs):
        print(f"Epoch {epoch+1}/{num_epochs}")
        print("-" * 20)

        train_loss = 0.0
        valid_loss = 0.0

        model_ = model_.to(device)

        # Обучение модели на обучающем наборе
        model_.train()
        print(f"Train {epoch+1}")
        for inputs, labels in dataloaders['train']:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer_.zero_grad()
            outputs = model_(inputs)
            loss = criterion_(outputs, labels)
            loss.backward()
            optimizer_.step()
            train_loss += loss.item() * inputs.size(0)
            # Работа с метриками
            for metric in metrics.values():
                metric(outputs, labels)

        train_loss = train_loss / len(dataloaders['train'].dataset)
        train_loss_history.append(train_loss)

        print(f"Train Loss: {train_loss:.4f}")

        for name_m, metric in metrics.items():
            t_m_c = metric.compute()
            train_metrics_history.get(name_m).append(t_m_c.cpu())
            print(f"{name_m}: {t_m_c}")

        for metric in metrics.values():
            metric.reset()

        # Проверка модели на валидационном наборе
        model_.eval()
        print(f"Valid {epoch+1}")
        with torch.no_grad():
            for inputs, labels in dataloaders['valid']:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model_(inputs)
                loss = criterion_(outputs, labels)
                valid_loss += loss.item() * inputs.size(0)
                # Работа с метриками
                for metric in metrics.values():
                    metric(outputs, labels)

            valid_loss = valid_loss / len(dataloaders['valid'].dataset)
            valid_loss_history.append(valid_loss)

            print(f"Valid Loss: {valid_loss:.4f}")

            for name_m, metric in metrics.items():
                v_m_c = metric.compute()
                valid_metrics_history.get(name_m).append(v_m_c.cpu())
                print(f"{name_m}: {v_m_c}")

            for metric in metrics.values():
                metric.reset()

learn(resnet, optimizer, criterion)
```

#### Модификация модели

Модель имеет ряд проблем:
  * Переобучение на один из классов
  * Схожесть классов geo и cl, а так же классов geo и kr
  * Проблема переобучения модели в целом и её адаптации к тренировочным данным

В основном эти проблемы вызваны несбалансированным дата-сетом. Для решения проблемы схожести классов и дисбаланса в дата-сете, можно применить взвешивание потерь, чтобы модель больше отдавала предпочтение классам cl и kr, между которами модель видит явные различия. А так же понизить скорость обучения модели для предотвращения адаптации к тренировочным данным.

Рассчитываем веса:

```python
num_samples = len(dataloaders['train'].dataset)
class_freqs = [0] * len(dataloaders['train'].dataset.classes)
for data in dataloaders['train'].dataset:
    class_freqs[data[1]] += 1
class_weights = [num_samples / (len(dataloaders['train'].dataset.classes) * freq) for freq in class_freqs]

weights = [class_weights[data[1]] for data in dataloaders['train'].dataset]
sampler = WeightedRandomSampler(weights, num_samples)

dataloaders.update({
    'train': DataLoader(image_datasets['train'], batch_size=batch_size, sampler=sampler)
})

class_weights
```

И так финальные оптимизатор и функция потерь примут вид:

```python
criterion = nn.CrossEntropyLoss(weight=torch.FloatTensor(class_weights)).to(device)
optimizer = optim.Adam(resnet.parameters(), lr=0.0002)
```

#### Проверка модели ResNet50 на тестовых данных

```python
def test(model_):
    global dataloaders, metric

    clear_metric()

    model_.eval()
    model_ = model_.to(device)

    with torch.no_grad():
        for inputs, labels in dataloaders['test']:
            print(labels)
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model_(inputs)
            _, predicted = torch.max(outputs, 1)
            print(predicted)

            for metric in metrics.values():
                metric(outputs, labels)

        for name, metric in metrics.items():
            if name == 'confusion_matrix':
                metric = metric.compute().cpu()
                fig, ax = plt.subplots()
                im = ax.matshow(metric, cmap=plt.cm.Blues)
                ax.set_xticks(np.arange(len(metric)))
                ax.set_yticks(np.arange(len(metric)))
                ax.set_xticklabels(['cl', 'geo', 'kr'])
                ax.set_yticklabels(['cl', 'geo', 'kr'])
                ax.xaxis.set_label_position('bottom')
                ax.xaxis.tick_bottom()
                for i in range(len(metric)):
                    for j in range(len(metric)):
                        ax.text(j, i, format(metric[i, j], 'd'), ha="center", va="center", color='white' if metric[i, j] > metric.max() / 2 else 'black')
                plt.xlabel('Predicted Label')
                plt.ylabel('True Label')
                plt.title('Confusion Matrix')
                plt.colorbar(im)
            else:
                print(f"{name}: {metric.compute().cpu()}")

        for _, metric in metrics.items():
            metric.reset()
test(resnet)
```
